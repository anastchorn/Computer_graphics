```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, clear_output

def equation_1(params, iterations, color, M=10000):
    x, y = 0, 0
    x_vals, y_vals = [], []
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.axis('off')  
    
    for i in range(iterations // M):
        for _ in range(M):
            p = params[np.random.choice(len(params))]
            next_x = p[0]* x + p[1] * y + p[4]
            next_y = p[2] * x + p[3] * y + p[5]
            x, y = next_x, next_y
            x_vals.append(x)
            y_vals.append(y)
        
        ax.clear()  
        ax.scatter(x_vals, y_vals, s=0.1, color=color)
        ax.axis('off')  
        display(fig)  
        
        clear_output(wait=True) 
        plt.pause(0.01)  

    plt.close(fig)  

def equation_2(params, iterations, probabilities, color, M=10000):
     x, y = 0, 0
     x_vals, y_vals = [], []
     fig, ax = plt.subplots(figsize=(6, 6))
     ax.axis('off')  
     for i in range(iterations // M):
           for _ in range(M):
               p = np.random.choice(len(params), p=probabilities)
               params_selected = params[p]
               next_x = params_selected[0] * x + params_selected[1] * y + params_selected[4]
               next_y = params_selected[2] * x + params_selected[3] * y + params_selected[5]
               x, y = next_x, next_y
               x_vals.append(x)
               y_vals.append(y)
           ax.clear() 
           ax.scatter(x_vals, y_vals, s=0.1, color=color)
           ax.axis('off')  
           display(fig)  
           clear_output(wait=True)  
           plt.pause(0.01)  

def equation_3(params, iterations, color, M=10000):
     x, y = 0, 0
     x_vals, y_vals = [], []
     fig, ax = plt.subplots(figsize=(6, 6))
     ax.axis('off') 
     for i in range(iterations // M):
           for _ in range(M):
               r, s, theta, phi, e, f = params[np.random.choice(len(params))]
               next_x = r * np.cos(theta) * x - s * np.sin(phi) * y + e
               next_y = r * np.sin(theta) * x + s * np.cos(phi) * y + f
               x, y = next_x, next_y
               x_vals.append(x)
               y_vals.append(y)
           ax.clear()
           ax.scatter(x_vals, y_vals, s=0.1, color=color)
           ax.axis('off')  
           display(fig) 
           clear_output(wait=True) 
           plt.pause(0.01)  
```
