```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def equation_1(params, iterations):
    x, y = 0, 0
    points = []
    for _ in range(iterations):
      p = params[np.random.choice(len(params))]
      next_x = p[0] * x + p[1] * y + p[4]
      next_y = p[2] * x + p[3] * y + p[5]
      x, y = next_x, next_y
      points.append((x, y))
    return points

def equation_2(params,iterations, probabilities):
    x, y = 0, 0
    points = []
    for _ in range(iterations):
      p = np.random.choice(len(params), p=probabilities)
      params_selected = params[p]
      next_x = params_selected[0] * x + params_selected[1] * y + params_selected[4]
      next_y = params_selected[2] * x + params_selected[3] * y + params_selected[5]
      x, y = next_x, next_y
      points.append((x, y))
    return points

def equation_3(params, iterations):
    x, y = 0, 0
    points = []
    for _ in range(iterations):
      r, s, theta, phi, e, f = params[np.random.choice(len(params))]
      next_x = r * np.cos(theta) * x - s * np.sin(phi) * y + e
      next_y = r * np.sin(theta) * x + s * np.cos(phi) * y + f
      x, y = next_x, next_y
      points.append((x, y))
    return points
```

```python
def draw_fractal(points, color):
    x_vals, y_vals = zip(*points)
    plt.figure(figsize=(6, 6))
    plt.scatter(x_vals, y_vals, s=0.1, color=color)
    plt.axis('off') 
    plt.show()
```

```python
params_for_Mandelbrotlike = [
    (0.2020, -0.8, -0.5, -0.3, 0.5, 0.5),
    (0.1080, 0.6650, -0.4, -0.2, 0.5, -0.1)
]
Mandelbrotlike_points=equation_1(params_for_Mandelbrotlike, 100000)
draw_fractal(Mandelbrotlike_points, 'black')
```

```python
params_for_fractal_leaf = [
    (0.0000, 0.2439, 0.0000, 0.3053, 0.0000, 0.0000),
    (0.7248, 0.1237, 0.0253, 0.7426, 0.2060, 0.2538),
    (0.1583, 0.1297, 0.3550, 0.3676, 0.1383, 0.1750),
    (0.3386, 0.3694, 0.2227, 0.1756, 0.0679, 0.0826)
]
fractal_leaf_points=equation_1(params_for_fractal_leaf, 100000)
draw_fractal(fractal_leaf_points, 'orange')
```

```python
params_for_fern_fractal = [
    (0.1, 0.0, 0.0, 0.16, 0.0, 0.0),
    (0.1, -0.26, 0.23, 0.22, 0.0, 1.6),
    (-0.1, 0.28, 0.26, 0.24, 0.0, 0.44),
    (0.65, 0.04, -0.04, 0.85, 0.0, 1.6)
]
probabilities_for_fern_fractal=[0.05, 0.05, 0.1, 0.8]
fern_fractal_points=equation_2(params_for_fern_fractal, 100000, probabilities_for_fern_fractal)
draw_fractal(fern_fractal_points, 'green')
```
