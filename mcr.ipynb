```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, clear_output
```

```python
#Варіант 9
def generate_points_stepwise(base=5, digits_to_exclude={0, 4}, depth=5, iterations_per_step=10000):
    points = []
    fig, ax = plt.subplots(figsize=(6, 6))
    
    x_vals, y_vals = [], []
    
    for x in range(base**depth):
        for y in range(base**depth):
            x_digits = [int(d) for d in np.base_repr(x, base=base).zfill(depth)]
            y_digits = [int(d) for d in np.base_repr(y, base=base).zfill(depth)]
            
            if not any(d in digits_to_exclude for d in x_digits) and not any(d in digits_to_exclude for d in y_digits):
                x_coord = sum(d * (base ** -i) for i, d in enumerate(x_digits, start=1))
                y_coord = sum(d * (base ** -i) for i, d in enumerate(y_digits, start=1))
                x_vals.append(x_coord)
                y_vals.append(y_coord)
                
                if len(x_vals) % iterations_per_step == 0:
                    ax.clear()
                    ax.scatter(x_vals, y_vals, s=0.1, color='blue')
                    ax.set_title(f'Fractal with base {base} excluding digits {digits_to_exclude}')
                    ax.set_xlabel('x')
                    ax.set_ylabel('y')
                    display(fig)
                    clear_output(wait=True)
                    plt.pause(0.01)
                    
    return x_vals, y_val
```
